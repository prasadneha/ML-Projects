```python
# import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#import the dataset
dataset=pd.read_csv("diabetes_prediction_dataset.csv")
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
# checking for any missing values
dataset.isnull().sum()
```

```python
dataset.describe()
```

```python
## correlation plot of independent variable
plt.figure(figsize=(10,8))
sns.heatmap(dataset.corr(),annot=True,fmt='.3f',cmap='YlGnBu')
plt.title("Correlation Heatmap")
```

```python
#Exploring age and diabetes variable
plt.figure(figsize=(10,8))
# plotting density function grapgh of the ages and diabetes variable
kde=sns.kdeplot(dataset['age'][dataset['diabetes']==1],color='Red',shade=True)
kde=sns.kdeplot(dataset['age'][dataset['diabetes']==0],color='green',shade=True)
kde.set_xlabel("Ages")
kde.set_ylabel("Density")
kde.legend(['Positive','Negative'])
```
